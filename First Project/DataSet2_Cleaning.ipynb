# Measure of central tendency & Dispersion

```python
for col in dt.select_dtypes(include=[np.number]).columns:
    print(f"\n--- {col} ---")
    print("Mean:", dt[col].mean())
    print("Median:", dt[col].median())
    print("Mode:", dt[col].mode()[0])
    print("Variance:", dt[col].var())
    print("Standard Deviation:", dt[col].std())
    print("Range:", dt[col].max() - dt[col].min())
    print("Skewness:", dt[col].skew())
    print("Kurtosis:", dt[col].kurt())
```

# Exploratory Data Analysis (EDA)

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
for col in dt.select_dtypes(include=[np.number]).columns:
    plt.figure(figsize=(6,4))
    sns.histplot(dt[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
for col in dt.select_dtypes(include=[np.number]).columns:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=dt[col], color='blue')
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
#Missing values heatmap
plt.figure(figsize=(6,2))
sns.heatmap(dt.isnull(), cbar=False)
plt.title("Missing values")
plt.show()
```

```python
plt.plot(dt['Date'], dt['Price_WTI'], color='green')
plt.title("Price_WTI Over Time")
plt.xlabel("Date")
plt.ylabel("Price_WTI")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
for col in dt.select_dtypes(include=[np.number]).columns:
    plt.figure(figsize=(12,5))
    sns.scatterplot(x=dt[col], y='Date', data=dt)
    plt.title(f"Scatterplot of {col}")
    plt.show()
```

```python
for col in dt.select_dtypes(include=[np.number]).columns:
    plt.figure(figsize=(10,6))
    sns.boxplot(x=dt[col])
    plt.title(f"Box plot of {col}")
    plt.show()
```

```python
# Time-series plot of close price
plt.figure(figsize=(12,5))
plt.plot(dt['Date'], dt['Price_All_Share'], color='red')
plt.title("Price_All_Share Over Time")
plt.xlabel("Date")
plt.ylabel("Price_All_Share")
plt.show()
```

```python
sns.heatmap(dt.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.pairplot(dt[['Price_WTI', 'Price_All_Share', 'Price_ICE']])
plt.suptitle("Pairplot of Numeric Features", y=1.02)
plt.show()
```
